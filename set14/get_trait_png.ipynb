```python
from bs4 import BeautifulSoup
import requests
from collections import defaultdict
import re
import pandas as pd
import os
```

```python
# Klassen
url = "https://tftactics.gg/db/classes/"
res = requests.get(url)
comm = re.compile("<!--|-->")

# BeautifulSoup zum Parsen nutzen
soup = BeautifulSoup(comm.sub("",res.text),'html')
data = soup.find_all("div", class_ = "rt-table")
data = data[0].find_all("div", class_ = "rt-tbody")
data = data[0].find_all("div", class_ = "character-wrapper")

class_list = []
for png in data:
    img_tag = png.find('img')  # Das <img> Tag finden
    if img_tag and 'src' in img_tag.attrs:
        link = img_tag['src']
        class_list.append(link)

icon_class_links = [link for link in class_list if "icon" in link]
class_names = []
for url in icon_class_links:
    # Die URL nach '/' teilen und den letzten Teil (den Namen) extrahieren
    name = url.split('/')[-1].split('.')[0].capitalize()
    class_names.append(name)

folder_name = 'downloaded_classes'

# Ordner erstellen, falls er noch nicht existiert
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Bilder herunterladen und speichern
for i in range(len(icon_class_links)):
    try:
        # URL und Name extrahieren
        url = icon_class_links[i]
        class_name = class_names[i]
        
        # Herunterladen des Bildes
        response = requests.get(url)
        
        # Überprüfen, ob die Anfrage erfolgreich war (Statuscode 200)
        if response.status_code == 200:
            # Bilddateiname mit dem Namen aus der Liste (z.B. 'Akali.png')
            file_name = os.path.join(folder_name, f"{class_name}.png")
            
            # Bild speichern
            with open(file_name, 'wb') as file:
                file.write(response.content)
            #print(f"Bild '{character_name}.png' erfolgreich heruntergeladen und gespeichert!")
        else:
            print(f"Fehler beim Herunterladen von {class_name}")
    
    except Exception as e:
        print(f"Fehler bei Bild {class_name}: {e}")
```
